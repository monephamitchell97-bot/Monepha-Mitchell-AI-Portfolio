## 2. Understanding Agent Frameworks

### What is an Agent Framework?

An agent framework is a set of pre-built components and tools that help you build, test, and deploy AI agents. Instead of writing everything from scratch, you can use the framework's components to quickly assemble a powerful agent. Popular agent frameworks include LangChain, LlamaIndex, and AutoGen.

### Key Components of an Agent

Most agent frameworks are built around three key concepts:

1.  **Tools:** These are the specific actions the agent can take. A tool could be a function that searches the web, a calculator, a database lookup, or even another AI model.
2.  **Memory:** This is where the agent stores information it has learned. Memory can be short-term (remembering the last few things that happened) or long-term (storing information for future use).
3.  **Workflow (or Chain):** This is the logic that the agent follows to decide which tool to use and when. It's the 'brain' of the agent that orchestrates the entire process.

## Part 1: Coded Demonstration

In this part, we will build a simple agent that can answer questions about the weather. We will create our own mini-framework to understand the core concepts.

```python
# 1. Define the Agent's Tools

def get_weather(city):
    """A tool that returns the weather for a given city."""
    if city == "Boston":
        return "The weather in Boston is 70 degrees and sunny."
    elif city == "San Francisco":
        return "The weather in San Francisco is 60 degrees and foggy."
    else:
        return f"Sorry, I don't know the weather in {city}."

def get_population(city):
    """A tool that returns the population of a given city."""
    if city == "Boston":
        return "The population of Boston is about 675,000."
    elif city == "San Francisco":
        return "The population of San Francisco is about 875,000."
    else:
        return f"Sorry, I don't know the population of {city}."
```

```python
# 2. Define the Agent's Workflow (the 'brain')

def weather_agent(prompt):
    """A simple agent that can answer questions about weather and population."""

    # Simple reasoning: look for keywords in the prompt
    if "weather" in prompt:
        # Extract the city from the prompt
        if "Boston" in prompt:
            city = "Boston"
        elif "San Francisco" in prompt:
            city = "San Francisco"
        else:
            return "I can only get the weather for Boston and San Francisco."

        # Use the get_weather tool
        return get_weather(city)

    elif "population" in prompt:
        # Extract the city from the prompt
        if "Boston" in prompt:
            city = "Boston"
        elif "San Francisco" in prompt:
            city = "San Francisco"
        else:
            return "I can only get the population for Boston and San Francisco."

        # Use the get_population tool
        return get_population(city)

    else:
        return "I'm sorry, I can only answer questions about weather or population."
```

```python
# 3. Use the agent

prompt1 = "What is the weather in Boston?"
response1 = weather_agent(prompt1)
print(f"Prompt: {prompt1}")
print(f"Response: {response1}")

prompt2 = "What is the population of San Francisco?"
response2 = weather_agent(prompt2)
print(f"Prompt: {prompt2}")
print(f"Response: {response2}")

prompt3 = "What is the capital of Texas?"
response3 = weather_agent(prompt3)
print(f"Prompt: {prompt3}")
print(f"Response: {response3}")
```
